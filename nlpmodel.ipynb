```python
for pos in range(data_cleaned_copy.shape[0]):
    for tuple_element in data_cleaned_copy['Extracted_Symptoms_copy'].iloc[pos]:
        ex_list = tuple_element[0].split()
        count_test = 0
        count_treat = 0
        count_prob = 0
        for ex_word in ex_list:
            try:
                ex_word_pos = data_cleaned_copy['SYMPTOM_TEXT_tokenized'].iloc[pos].index(ex_word)
                if tuple_element[1] == 'TEST':
                    if count_test == 0:
                        data_cleaned_copy['Test_Tags'].iloc[pos][ex_word_pos] = 1
                        count_test += 1
                    else:
                        data_cleaned_copy['Test_Tags'].iloc[pos][ex_word_pos] = 2
                elif tuple_element[1] == 'TREATMENT':
                    if count_treat == 0:
                        data_cleaned_copy['Treatment_Tags'].iloc[pos][ex_word_pos] = 1
                        count_treat += 1
                    else:
                        data_cleaned_copy['Treatment_Tags'].iloc[pos][ex_word_pos] = 2
                else:  # PROBLEM set
                    if count_prob == 0:
                        data_cleaned_copy['Problem_Tags'].iloc[pos][ex_word_pos] = 1
                        count_prob += 1
                    else:
                        data_cleaned_copy['Problem_Tags'].iloc[pos][ex_word_pos] = 2
            except ValueError:
                # Handle the case where the word is not found in the tokenized list
                continue
```

```python
def encode(tokenizer, texts, tags):
    tokenized_inputs = tokenizer(texts, truncation=True, padding=True, is_split_into_words=True)
    labels = []

    for i, label in enumerate(tags):
        word_ids = tokenized_inputs.word_ids(batch_index=i)
        previous_word_idx = None
        label_ids = []

        for word_idx in word_ids:
            if word_idx is None:
                label_ids.append(-100)  # For special tokens
            elif word_idx != previous_word_idx:
                label_ids.append(label[word_idx] if word_idx < len(label) else -100)
            else:
                # For subwords that are not the first subword token
                label_ids.append(-100)  # or use same label as the first subword token if necessary

            previous_word_idx = word_idx

        labels.append(label_ids)

    tokenized_inputs["labels"] = labels
    return tokenized_inputs
```

```python
from datasets import load_metric
import numpy as np
```

```python
accuracy_metric = load_metric("accuracy")
precision_metric = load_metric("precision")
recall_metric = load_metric("recall")
f1_metric = load_metric("f1")
```

```python
def compute_metrics(eval_pred):
    logits, labels = eval_pred
    predictions = np.argmax(logits, axis=-1)
    # Flatten predictions and labels for metric calculation
    true_labels = labels.flatten()
    pred_labels = predictions.flatten()

    # Filter out special tokens (label == -100)
    true_labels = true_labels[true_labels != -100]
    pred_labels = pred_labels[labels.flatten() != -100]

    accuracy = accuracy_metric.compute(predictions=pred_labels, references=true_labels)
    precision = precision_metric.compute(predictions=pred_labels, references=true_labels, average='macro')
    recall = recall_metric.compute(predictions=pred_labels, references=true_labels, average='macro')
    f1 = f1_metric.compute(predictions=pred_labels, references=true_labels, average='macro')

    return {"accuracy": accuracy, "precision": precision, "recall": recall, "f1": f1}
```

```python
predictions_problem, label_ids_problem, metrics_problem = trainer_test.predict(data_Test_test_dataset)
metrics_calculated_problem = compute_metrics((predictions_problem, label_ids_problem))
```

```python
print("Metrics for Problem Model:", metrics_calculated_problem)
```
