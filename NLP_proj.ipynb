```python
sampled_reports['num_symptoms'] = sampled_reports[['SYMPTOM1', 'SYMPTOM2', 'SYMPTOM3', 'SYMPTOM4', 'SYMPTOM5']].notnull().sum(axis=1)

# Creating SYMPTOM_TEXT by concatenating all symptoms
sampled_reports['SYMPTOM_TEXT'] = sampled_reports.apply(lambda x: ' '.join(x[['SYMPTOM1', 'SYMPTOM2', 'SYMPTOM3', 'SYMPTOM4', 'SYMPTOM5']].dropna().astype(str)), axis=1)
sampled_reports['report_length'] = sampled_reports['SYMPTOM_TEXT'].apply(len)

# Group by vaccine type and get distribution of number of symptoms and report length
grouped_data = sampled_reports.groupby('VAX_NAME').agg({
    'num_symptoms': ['mean', 'std', 'min', 'max', 'count'],
    'report_length': ['mean', 'std', 'min', 'max', 'count']
})
```

```python
symptom_columns = ['SYMPTOM1', 'SYMPTOM2', 'SYMPTOM3', 'SYMPTOM4', 'SYMPTOM5']
standard_symptom_list = pd.concat([sampled_reports[col] for col in symptom_columns]).dropna()

# Finding the most common symptoms (the most frequent 100 symptoms)
most_common_100_symptoms = standard_symptom_list.value_counts().head(100)

most_common_100_symptoms
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

# Convert the most common symptoms to a string, with each symptom repeated according to its frequency
symptom_string = ' '.join([symptom for symptom in most_common_100_symptoms.index for _ in range(most_common_100_symptoms[symptom])])

# Create a WordCloud object
wordcloud = WordCloud(width=800, height=800, background_color='white', min_font_size=10).generate(symptom_string)

# Plot the WordCloud image
plt.figure(figsize=(8, 8), facecolor=None)
plt.imshow(wordcloud)
plt.axis("off")
plt.tight_layout(pad=0)

plt.show()
```

```python
import stanza

# Download the i2b2 model
stanza.download('en', package='mimic', processors={'ner': 'i2b2'})
```

```python
# Create a Stanza pipeline with the i2b2 NER model
nlp = stanza.Pipeline(lang='en', package='mimic', processors={'ner': 'i2b2'})
def extract_symptom_entities(text):
    # Process the text
    doc = nlp(text)
    # Extract entities labeled as PROBLEM (symptoms)
    #symptom_entities = [ent.text for sent in doc.sentences for ent in sent.ents if ent.type == 'PROBLEM']
    return [(ent.text, ent.type) for sent in doc.sentences for ent in sent.ents]
```

```python
# Example usage with a few reports
sample_texts = sampled_reports['SYMPTOM_TEXT'].head(20)  # Select the first 20 reports
extracted_entities = [extract_symptom_entities(text) for text in sample_texts]

extracted_entities
```

```python
# Apply the function to the SYMPTOM_TEXT column
sampled_reports['Extracted_Symptoms'] = sampled_reports['SYMPTOM_TEXT'].apply(extract_symptom_entities)
```
